```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline 
```

```python
df = pd.read_csv('nba_feb.csv')
```

```python
df1 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2020_games.html')
df2 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2020_games-november.html')
df3 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2020_games-december.html')
df4 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2020_games-january.html')
df5 = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2020_games-february.html')
```

```python
frames = [df1[0], df2[0], df3[0], df4[0], df5[0]]
```

```python
all_games = pd.concat(frames)
```

```python
all_games
```

```python
kings_games = all_games[(all_games['Home/Neutral'] =='Sacramento Kings') | (all_games['Visitor/Neutral'] =='Sacramento Kings')]
```

```python
kings_games.head(30)
```

```python
kings_games.loc[kings_games['Home/Neutral'] == 'Sacramento Kings', 'Notes'] = 'Home'
kings_games.loc[kings_games['Visitor/Neutral'] == 'Sacramento Kings', 'Notes'] = 'Away'
kings_games
```

```python
kings_games['W/L'] = 'N/A'
```

```python
kings_games['Kings Points'] = 0
kings_games['Opponent Points'] = 0
```

```python
kings_games
```

```python
kings_games = kings_games[~kings_games.index.duplicated()]

```

```python
kings_games.loc[kings_games['Home/Neutral'] == 'Sacramento Kings', 'Kings Points'] = kings_games['PTS.1']
                                                                                                 
kings_games.loc[kings_games['Visitor/Neutral'] == 'Sacramento Kings', 'Kings Points'] = kings_games['PTS'] 

kings_games.loc[kings_games['Home/Neutral'] != 'Sacramento Kings', 'Opponent Points'] = kings_games['PTS.1']
                                                                                                 
kings_games.loc[kings_games['Visitor/Neutral'] != 'Sacramento Kings', 'Opponent Points'] = kings_games['PTS'] 
```

```python
kings_games
```

```python
kings_games.loc[kings_games['Kings Points'] > kings_games['Opponent Points'], 'W/L'] = 'W'
kings_games.loc[kings_games['Kings Points'] < kings_games['Opponent Points'], 'W/L'] = 'L'
```

```python
kings_games.drop(kings_games.loc[kings_games['W/L']=='N/A'].index, inplace=True)
```

```python
sns.barplot(x='Notes',y='Kings Points',data=kings_games)
```

```python
sns.barplot(x='Notes',y='Opponent Points',data=kings_games)
```

```python
sns.set(style="darkgrid")
ax = sns.countplot(x='W/L',hue='Notes', data=kings_games)
```

```python
sns.violinplot(x='W/L', y = 'Kings Points',data=kings_games, hue='Notes', split = True)
```

```python
sns.violinplot(x='Notes', y = 'Kings Points',data=kings_games, hue='W/L', split = True)
```

```python
sns.set(style="whitegrid")


ax = sns.swarmplot(x='Notes', y = 'Kings Points',data=kings_games, hue='W/L', split = True)
ax.legend_out = False
ax
```

```python
g = sns.FacetGrid(data=kings_games, col='Notes', row= 'W/L')
g.map(sns.distplot, 'Kings Points')
```
